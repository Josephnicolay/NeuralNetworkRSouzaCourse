```python
import numpy as np
```

```python
x=np.linspace(-np.pi,np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,5))
ax.plot(x,y1,'--r',label='sin(x)')
ax.plot(x,y2,'-b',label='cos(x)')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('True')
fig.savefig('my_plot.png')
```
